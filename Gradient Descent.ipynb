## Gradient Descent

```python
# Gradient Descent example to minimize f(x) = x^2

def f(x):
    return x**2

def gradient(x):
    return 2 * x

def gradient_descent(start_x, learning_rate, iterations):
    x = start_x
    for i in range(iterations):
        grad = gradient(x)
        x = x - learning_rate * grad
        print(f"Iteration {i+1}: x = {x:.5f}, f(x) = {f(x):.5f}")
    return x

# Parameters
start_x = 10  # Starting point
learning_rate = 0.1  # Step size
iterations = 20  # Number of iterations

final_x = gradient_descent(start_x, learning_rate, iterations)
print(f"\nMinimum value found at x = {final_x:.5f}, f(x) = {f(final_x):.5f}")
```

## Complete Code

```python
import matplotlib.pyplot as plt

# Define the function f(x) = x^2
def f(x):
    return x ** 2

# Define the derivative (gradient) f'(x) = 2x
def gradient(x):
    return 2 * x

# Gradient descent function
def gradient_descent(start_x, learning_rate, iterations):
    x = start_x
    x_values = []
    f_values = []

    for i in range(iterations):
        grad = gradient(x)
        x = x - learning_rate * grad
        x_values.append(x)
        f_values.append(f(x))
        print(f"Iteration {i+1}: x = {x:.5f}, f(x) = {f(x):.5f}")

    return x_values, f_values

# Parameters
start_x = 10         # Starting point
learning_rate = 0.1  # Learning rate
iterations = 20      # Number of iterations

# Run gradient descent
x_vals, f_vals = gradient_descent(start_x, learning_rate, iterations)

# Plotting the results
plt.figure(figsize=(12, 5))

# Plot x values over iterations
plt.subplot(1, 2, 1)
plt.plot(range(1, iterations + 1), x_vals, marker='o')
plt.title('Value of x over iterations')
plt.xlabel('Iteration')
plt.ylabel('x')

# Plot f(x) values over iterations
plt.subplot(1, 2, 2)
plt.plot(range(1, iterations + 1), f_vals, marker='o', color='red')
plt.title('Value of f(x) over iterations')
plt.xlabel('Iteration')
plt.ylabel('f(x)')

plt.tight_layout()
plt.show()
```

#### The output will show the values of x and f(x) at each iteration printed to the console and two graphs displaying how x and f(x) change over the iterations.
